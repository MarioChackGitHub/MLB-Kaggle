```python
#Gus, cómo sacar json del csv train_updated
import json
import gc  #garbage collector

traindf = pd.read_csv("data/train_updated.csv", nrows=100)
traindf[['date']+["nextDayPlayerEngagement"]].dropna(how='any').reset_index(drop=True)
```

```python
#funcion de Gus
def unpack_json(df,var):
    name = var
    tmp = df[['date']+[name]].dropna(how='any').reset_index(drop=True)
    temp=[]
    for val in tmp[name]:
        temp.append(json.loads(val))
    tmp[name]=temp
    
    var_val_dict = dict(zip(tmp.iloc[0,1][0].keys(),[[] for val in tmp.iloc[0,1][0].keys()]))
    date_list = []

    for d, eng in zip(tmp['date'].values,tmp[name].values):
        for val in eng:
            date_list.append(d)
            for key in val.keys():
                var_val_dict[key].append(val[key])
                #var_val_dict['date_tr'].append(id1)
                #tqdm(zip(df['date'].values,df[name].values))

    output_df = pd.DataFrame(var_val_dict)
    output_df['date'] = date_list
    gc.collect()
    
    return output_df
```

```python
#seleccionado
targets_df = unpack_json(traindf,'nextDayPlayerEngagement')
rosters_df = unpack_json(traindf,'rosters')
#scores_tmp_df = unpack_json(train_df,'playerBoxScores')
#games_df = unpack_json(train_df,'games')
#teamBoxScores_df = unpack_json(train_df,'teamBoxScores')

players_df = pd.read_csv("data/players.csv")
#scores = scores_tmp.groupby(['playerId', 'date']).sum().reset_index()
teams_df = pd.read_csv('data/teams.csv')
seasons_df = pd.read_csv('data/seasons.csv')
```

```python
targets_df_cols = ['playerId', 'target1', 
    'target2', 'target3', 'target4', 'date']
rosters_df_cols = ['playerId', 'teamId', 
    'status', 'date']
players_df_cols = ['playerId', 'DOB', 
    'primaryPositionName', 'mlbDebutDate']
```

## hacer join de los dfs por id players!

```python
train_df = targets_df[targets_df_cols].merge(players_df[players_df_cols], 
                                            on = ["playerId"], how = "left")\
                                        .merge(rosters_df[rosters_df_cols],
                                            on = ["playerId", "date"], how="left")\
                                        .merge(teams_df, left_on= ["teamId"], right_on=["id"], how="left")
print("0", train_df.shape)

train_df.head()
```

```python
#este es nuestro df final para hacer training de los datos pero tenemos que formattear la fecha para un mejor performance.
print(train_df.shape)
train_df.head()
```

```python
#cambiamos el formato de la fecha:
train_df["year"] = pd.to_datetime(
                        train_df['date'], format = "%Y%m%d") \
                        .map(lambda x:str(x.year).zfill(4)).astype(str)
train_df["month"] = pd.to_datetime(
                        train_df['date'], format = "%Y%m%d") \
                        .map(lambda x:str(x.month).zfill(2)).astype(str)
train_df["day"] = pd.to_datetime(
                        train_df['date'], format = "%Y%m%d") \
                        .map(lambda x:str(x.day).zfill(2)).astype(str)
```
